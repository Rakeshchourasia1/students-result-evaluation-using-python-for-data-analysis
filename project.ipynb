```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("students_details.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop("Unnamed: 0", axis=1)
print(df.head())
```

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above graph we actually know how much girls and boys are in the school
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

```python
#from the above heat map we conclude that the parents education qualification throws a good impact on the students marks
```

```python
gp=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gp)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gp,annot=True)
plt.title("Relationship between Parent's maritial status and Student's Score")
plt.show()
```

```python
#from the above chart we conclude that the parents marital status did not effect too much on the students result.
```

```python
#Distribution of Ethnic groups
groupA=df.loc[(df['EthnicGroup']=='group A')].count()
groupB=df.loc[(df['EthnicGroup']=='group B')].count()
groupC=df.loc[(df['EthnicGroup']=='group C')].count()
groupD=df.loc[(df['EthnicGroup']=='group D')].count()
groupE=df.loc[(df['EthnicGroup']=='group E')].count()

l=["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

plt.pie(mlist, labels=l,autopct="%1.2f%%")
plt.title("Distribution of Ethin Groups")
plt.show()
```
